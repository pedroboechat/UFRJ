```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import os
os.makedirs("./plots")
```

```python
c_ij = pd.read_csv("./output/output_c_ij.csv")
c_ji = pd.read_csv("./output/output_c_ji.csv")
fortran_ij = pd.read_csv("./output/output_fortran_ij.csv")
fortran_ji = pd.read_csv("./output/output_fortran_ji.csv")
```

```python
plt.plot(c_ij["n"], c_ij["elapsed"])
plt.plot(c_ji["n"], c_ji["elapsed"])
plt.xticks(np.arange(10000, 110001, 20000))
plt.title(f"C implementation performance")
plt.xlabel('Number of elements')
plt.ylabel('Time (seconds)')
plt.legend(["ij loop", "ji loop"])
plt.savefig(f"./plots/c_benchmark.jpg", dpi=300.0);
```

```python
plt.plot(c_ij["n"], round((c_ji["elapsed"]/c_ij["elapsed"])*100,2), color="orange")
plt.xticks(np.arange(10000, 110001, 20000))
plt.title(f"ij and ji loop comparison in C\n(ji over ij)")
plt.xlabel('Number of elements')
plt.ylabel('Time difference (%)')
plt.legend(["ji / ij"])
plt.savefig(f"./plots/c_comparison.jpg", dpi=300.0);
```

```python
plt.plot(fortran_ij["n"], fortran_ij["elapsed"])
plt.plot(fortran_ji["n"], fortran_ji["elapsed"])
plt.xticks(np.arange(10000, 110001, 20000))
plt.title(f"Fortran implementation performance")
plt.xlabel('Number of elements')
plt.ylabel('Time (seconds)')
plt.legend(["ij loop", "ji loop"])
plt.savefig(f"./plots/fortran_benchmark.jpg", dpi=300.0);
```

```python
plt.plot(fortran_ij["n"], round((fortran_ij["elapsed"]/fortran_ji["elapsed"])*100,2))
plt.xticks(np.arange(10000, 110001, 20000))
plt.title(f"ij and ji loop comparison in Fortran\n(ij over ji)")
plt.xlabel('Number of elements')
plt.ylabel('Time difference (%)')
plt.legend(["ij / ji"])
plt.savefig(f"./plots/fortran_comparison.jpg", dpi=300.0);
```
