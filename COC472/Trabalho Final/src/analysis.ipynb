## **Imports**

```python
import pandas as pd
import matplotlib.pyplot as plt
from os import makedirs
makedirs("./plots", exist_ok=True)
```

## **Carrega dados**

```python
df = pd.read_csv("./data/run.csv", sep=";")
df.head(1)
```

## **An√°lise da performance com diferentes NB**
### **1) Para N = 5040**

```python
fig, axs = plt.subplots(2, 2, figsize=(15,12))
axs[0,0].plot(list(df["NB"].unique()), list(df.loc[(df["N"] == 5040) & (df["P"] == 1)]["gflops"]), label="Gflops", linewidth=3, color="C0")
axs[0,0].set_title("Performance por NB\nN: 5040, P: 1")
axs[0,0].set(xlabel='NB', ylabel='Performance (Gflops)')
axs[0,1].plot(list(df["NB"].unique()), list(df.loc[(df["N"] == 5040) & (df["P"] == 1)]["time"]), label="Time (s)", linewidth=3, color="C1")
axs[0,1].set_title("Tempo de execu√ß√£o por NB\nN: 5040, P: 1")
axs[0,1].set(xlabel='NB', ylabel='Tempo de execu√ß√£o (s)')
axs[1,0].plot(list(df["NB"].unique()), list(df.loc[(df["N"] == 5040) & (df["P"] == 16)]["gflops"]), label="Gflops", linewidth=3, color="C0")
axs[1,0].set_title("Performance por NB\nN: 5040, P: 16")
axs[1,0].set(xlabel='NB', ylabel='Performance (Gflops)')
axs[1,1].plot(list(df["NB"].unique()), list(df.loc[(df["N"] == 5040) & (df["P"] == 16)]["time"]), label="Time (s)", linewidth=3, color="C1")
axs[1,1].set_title("Tempo de execu√ß√£o por NB\nN: 5040, P: 16")
axs[1,1].set(xlabel='NB', ylabel='Tempo de execu√ß√£o (s)')
for ax in axs.flat:
    ax.legend()
fig.subplots_adjust(hspace=0.4, wspace=0.2)
plt.rcParams.update({'font.size': 16})
plt.setp(axs[:,0], xticks=list(range(0,251,50))[1:], yticks=range(0, 11, 2));
plt.setp(axs[:,1], xticks=list(range(0,251,50))[1:], yticks=range(0, 81, 10));
plt.savefig(f"./plots/analysis_nb_5k.jpg", dpi=300.0);
```

## **An√°lise de desempenho com diferentes N**
### **1) Compara√ß√£o de performance**

```python
plt.figure(figsize=(10,5))
plt.bar([str(i) for i in df["N"].unique()], list(df.loc[(df["P"] == 1) & (df["NB"] == 256)]["gflops"]), color="C0")
plt.plot([str(i) for i in df["N"].unique()], list(df.loc[(df["P"] == 1) & (df["NB"] == 256)]["gflops"]), color="C1", linewidth=3)
plt.xlabel("N")
plt.ylabel("Performance (Gflops)")
plt.title("Performance por N\nNB: 256, P: 1")
plt.savefig(f"./plots/performance_n.jpg", dpi=300.0);
```
